```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread('assets/pic.jpg',)
cv2_imshow(img)
```

```python
print(img.shape)
```

```python
height, width, _ = img.shape
print(width)
print(height)
```

```python
hsv= cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

```python
lower_blu=np.array([90, 50, 50])
```

```python
upper_blue = np.array([130,255,255])
```

```python
mask = cv2.inRange(hsv, lower_blu, upper_blue)
```

```python
result = cv2.bitwise_and(img, img, mask=mask)
```

```python
cv2_imshow(result)
```
