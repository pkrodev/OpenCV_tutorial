```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('assets/board.png')
```

```python
img = cv2.resize(img, (0,0), fx=2.2, fy=2.2)
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_gray.shape
```

```python
corners = cv2.goodFeaturesToTrack(img_gray, 100, 0.00000001, 10)
```

```python
corners = np.intp(corners)
```

```python
print(corners)
```

```python
for corner in corners:
  x, y = corner.ravel() #[[1,2],[2,1]] ---> [1,2,2,1]
  cv2.circle(img,(x,y), 5, (255,0,0), -1)
```

```python
cv2_imshow(img)
```

```python
for i in range (len(corners)):
  for j in range(i+1, len(corners)):
    corner1 = tuple(corners[i][0])  #it's because we haven't used ravel()
    corner2 = tuple(corners[j][0])
    color=tuple(map(lambda x: int(x), np.random.randint(0,255, size=3)))
    cv2.line(img, corner1, corner2, color, 1)
```

```python
cv2_imshow(img)
```
