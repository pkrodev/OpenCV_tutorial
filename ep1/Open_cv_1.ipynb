```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('assets/pic.jpg',)
img = cv2.resize(img, (400,300)) #Fixed image size assignment
```

```python
cv2_imshow(img)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('assets/pic.jpg',)
img = cv2.resize(img, (0,0), fx=2, fy=2) #image scaling
```

```python
cv2_imshow(img)
cv2.destroyAllWindows()
```

```python
img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(img)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('new_image.jpg', img)
```
