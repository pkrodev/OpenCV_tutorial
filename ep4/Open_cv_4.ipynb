```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread('assets/pic.jpg',)
cv2_imshow(img)
```

```python
print(img.shape)
```

```python
height, width, _ = img.shape
print(width)
print(height)
```

```python
modified_img= cv2.line(img, (0,0), (width,height), (255,0,0),10) # image, starting position, ending position, color, line thickness
cv2_imshow(img)
```

```python
modified_img= cv2.line(img, (0,height), (width,0), (255,0,0),10)
cv2_imshow(img)
```

```python
img = cv2.rectangle(img, (100 , 100),(500,500), (0,0,0), -1)
cv2_imshow(img)
```

```python
img = cv2.circle(img, (700,150), 50, (0,0,255),-1)
cv2_imshow(img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
img=cv2.putText(img, 'Text test', (494, height-50), font, 2, (0,0,0), 2, cv2.LINE_AA)
cv2_imshow(img)
```
