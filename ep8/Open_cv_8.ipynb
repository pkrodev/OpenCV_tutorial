```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread('assets/pic.jpg', 0)
```

```python
cv2_imshow(img)
```

```python
template = cv2.imread('assets/face1.JPG', 0)
```

```python
cv2_imshow(template)
```

```python
h, w = template.shape
```

```python
methods = [cv2.TM_CCOEFF, cv2.TM_CCOEFF_NORMED, cv2.TM_CCORR, cv2.TM_CCORR_NORMED, cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]
```

```python
for method in methods:
  img2 = img.copy()

  result = cv2.matchTemplate(img2, template, method)
  min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
  print(min_loc, max_loc)
```

```python
for method in methods:
  img2 = img.copy()

  result = cv2.matchTemplate(img2, template, method)
  min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
  if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:   #those methods are looking for minimum values of difference
    location = min_loc
  else:
    location=max_loc
  bottom_right= (location[0] + w, location[1]+h)
  cv2.rectangle(img2, location, bottom_right, 0, 5)
  cv2_imshow(img2)
```
